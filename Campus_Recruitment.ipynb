##Splitting to Train and Test sets

```
from sklearn.model_selection import train_test_split
```

```
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.15)
```

```
print(len(x_train))
print(len(x_test))
print(len(y_train))
print(len(y_test))
```

##Label Encoder 

```
from sklearn.preprocessing import LabelEncoder 
```

```
le = LabelEncoder()
```

```
labeltrain  = le.fit_transform(y_train)
```

```
labeltrain
```

```
y_train
```

```
labeltest = le.fit_transform(y_test)
```

```
labeltest
```

```
y_test
```

##Standardization

```
from sklearn.preprocessing import StandardScaler
```

```
sc = StandardScaler()
```

```
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

##Train on Different Algorithms


```
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
```

```
l_cla = LogisticRegression()
k_cla = KNeighborsClassifier()
d_cla = DecisionTreeClassifier()
r_cla = RandomForestClassifier()
s_cla = SVC(kernel = 'linear')
ks_cla = SVC(kernel = 'rbf')
```

```
l_cla.fit(x_train, labeltrain)
k_cla.fit(x_train, labeltrain)
d_cla.fit(x_train, labeltrain)
r_cla.fit(x_train, labeltrain)
s_cla.fit(x_train, labeltrain)
ks_cla.fit(x_train, labeltrain)
```

```
l_pred = l_cla.predict(x_test)
k_pred = k_cla.predict(x_test)
d_pred = d_cla.predict(x_test)
r_pred = r_cla.predict(x_test)
s_pred = s_cla.predict(x_test)
ks_pred = ks_cla.predict(x_test) 
```

```
from sklearn.metrics import confusion_matrix
```

```
l_c = confusion_matrix(labeltest, l_pred)
k_c = confusion_matrix(labeltest, k_pred)
d_c = confusion_matrix(labeltest, d_pred)
r_c = confusion_matrix(labeltest, r_pred)
s_c = confusion_matrix(labeltest, s_pred)
ks_c = confusion_matrix(labeltest, ks_pred)
```

##Result

```
from sklearn.metrics import accuracy_score
```

```
l_a = accuracy_score(labeltest, l_pred)
k_a = accuracy_score(labeltest, k_pred)
d_a = accuracy_score(labeltest, d_pred)
r_a = accuracy_score(labeltest, r_pred)
s_a = accuracy_score(labeltest, s_pred)
ks_a = accuracy_score(labeltest, ks_pred)
```

```
print('Logistic Regression: ' + str(l_a) + '\nKNeighbors: ' + str(k_a) + '\nDecision Tree: ' + str(d_a) + '\nRandom Forest: ' + str(r_a) + '\nLinear SVC: ' + str(s_a) + '\nKernel SVC: ' + str(ks_a))
```
